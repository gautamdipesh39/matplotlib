```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, marker="o")
plt.title("Matplotlib Test")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

subjects = ["Math", "Science", "English", "Computer"]
marks = [85, 78, 92, 88]

plt.bar(subjects, marks, color="skyblue")
plt.title("Student Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.ylim(0, 100)

plt.show()
```

```python
import matplotlib.pyplot as plt

activities = ["Study", "Sleep", "Exercise", "Entertainment","Hobbies"]
hours = [8, 7, 2, 3,9]

plt.pie(hours, labels=activities, autopct="%1.0f%%")
plt.title("Daily Time Usage")

plt.show()
```

```python
import matplotlib.pyplot as plt

hours_studied = [1, 2, 3, 4, 5, 6]
exam_marks = [35, 45, 55, 65, 75, 85]

plt.scatter(hours_studied, exam_marks, color="green")
plt.title("Study Hours vs Exam Marks")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Marks")
plt.grid(True)

plt.show()
```
