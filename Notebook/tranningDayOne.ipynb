```python
#series from dictionary 
import pandas as pd 
grates={"semester1":3.12,
        "semester2":3.33,
        "semester3":3.44
        }

#create a series from the dictionary 
my_series = pd.Series(grates,index=("semester1","semester2","semester3"))

#Display the series 
print(my_series)
```

```python
#custom index puts semester3 first 
series_reorder = pd.Series(grates,index=("semester3","semester2","semester1"))
print(series_reorder)
```

```python
#leaving out 
series_reorder = pd.Series(grates,index=("semester3","semester2"))
print(series_reorder)
```

```python
import numpy as np
import pandas as pd

# Series A: indices a, b, c, d; b has a missing value
series_a = pd.Series([10, np.nan, 30, 40], index=["a", "b", "c", "d"])

# Series B: indices b, c, d; index a is missing
series_b = pd.Series([2, 3, 4], index=["b", "c", "d"])

standard_mul = series_a * series_b
filled_mul = series_a.mul(series_b, fill_value=1)

print("------ Standard multiplication ------")
print(standard_mul)

print("\n------ Multiplication with fill_value=1 ------")
print(filled_mul)
```

```python
import pandas as pd

df = pd.DataFrame({
    "name": ["aayush", "bina", "sulay", "sushil"],
    "city": ["kathmandu", "pokhara", "birathnagar", "Dangadi"],
    "Amount": [400, 500, 100, 300]
})

df1 = df.loc[1:2]  # label-based row selection
print(df1)
```

```python
print(df.loc[df["Amount"]>200])

```

```python
df.loc[df["Amount"]>100,"city"]="kathmandu"
print(df)
```

```python
print(df.iloc[0])
print("\n {o,2}")
print(df.iloc[0,2])
print("\n {3}")

print(df.iloc[3])
```

```python
#result = df[(df["Amount"]>100)) & (df["city"]=="kathmandu")]
result = df[(df["Amount"] > 100) & (df["city"]=="kathmandu")]
print(result)
print("\n ------another------ ")
result2 = df[(df["city"]=="kathmandu") | (df["city"]=="birathnagar")]
print(result2)
```
